```python
#Categorize the Reviews

# Define the reviews and labels
reviews = [-1, 0, 10, 50, 100, 200, 300, 538]
category = ['No Reviews', 'Very Few Reviews', 'Few Reviews', 'Moderate Reviews', 'Many Reviews', 'Very Many Reviews', 'Top Reviewed']

# Create a new column in the dataframe with the categorized reviews
df['reviews_category'] = pd.cut(df['reviews'], bins=reviews, labels=category)

# Display the first few rows to check
df[['reviews', 'reviews_category']].head()
```

```python
df.head(3)
```

```python
#Categorize the Overall Satisfaction

# Define the Satisfaction and labels
satisfaction = [-0.1, 1, 2, 3, 4, 4.5, 5]
category = ['Very Poor', 'Poor', 'Fair', 'Good', 'Very Good', 'Excellent']

# Create a new column in the dataframe with the categorized satisfaction levels
df['satisfaction_category'] = pd.cut(df['overall_satisfaction'],bins = satisfaction, labels = category)

# Display the first few rows to check
df[['overall_satisfaction', 'satisfaction_category']].head()
```

```python
df.head(3)
```

```python
#Categorize the different price

# Define the price and labels
price = [12, 100, 250, 500, 1000, 2000, 6000]
labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High', 'Luxury']

# Create a new column in the dataframe with the categorized price levels
df['price_category'] = pd.cut(df['price'], bins=price, labels=labels, include_lowest=True)

# Display the first few rows to check
df[['price', 'price_category']].head()
```

```python
df.head(3)
```

### Regarding the Host


#### Who are top earners

```python
df['total_earnings'] = df['price'] * df['reviews'] 
top_earners = df.groupby('host_id')['total_earnings'].sum().sort_values(ascending=False)
```

#### Is there any relationship between monthly earning and prices

```python
monthly_earnings = df.groupby('host_id').apply(lambda x: (x['price'] * x['reviews']).mean())
correlation = monthly_earnings.corr(df['price'])
```

### Regarding the Neighbourhood

#### Any particular location getting maximum number of bookings

```python
bookings_by_neighborhood = df.groupby('neighborhood')['reviews'].sum().sort_values(ascending=False)
```

#### Price relation with respect to location

```python
price_by_neighborhood = df.groupby('neighborhood')['price'].mean().sort_values(ascending=False)
```

#### Price vs location

```python
# Here I am showing top 3 because this graph is not showing properly

median_prices = df.groupby('neighborhood')['price'].median().reset_index()

# Sort neighborhoods by median price in descending order and get the top 3
top_3_neighborhoods = median_prices.sort_values(by='price', ascending=False).head(3)

print("Top 3 neighborhoods based on median price:")
print(top_3_neighborhoods)

# Optional: Create a boxplot to visualize the data
sns.boxplot(x='neighborhood', y='price', data=df)
plt.xticks(rotation=90)
plt.show()
```
